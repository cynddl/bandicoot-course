# Quick introduction to matplotlib

```python
# This line configures matplotlib to show figures embedded in the notebook, 
# instead of opening a new window for each figure.
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = range(10)
y = [i * i for i in x]
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x, y, 'go-')
```

```python
# Create two subplots side by side
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
plt.tight_layout() # Add spacing between graphs
```

```python
plt.figure(figsize=(12,6))
plt.plot(x, y, '.-')

# Use a log scale for the ordinate
plt.yscale('log')
```

## Saving the figures

matplotlib automatically detects the type of files (PDF, PNG…).

```python
plt.plot(x, y)
plt.savefig("filename.png")
```
